```python
import pandas as pd
import numpy as np

# 1) Read the Excel file
df = pd.read_excel("../data/e_redes_perfil_consumo.xlsx")

# 2) Set header from 3rd row and clean
df.columns = df.iloc[2]
df = df.drop([0, 1, 2]).reset_index(drop=True)

# 3) Choose columns: 1st NaN (date), 2nd NaN (hour) + valid ones
is_nan = pd.isna(df.columns)
first_nan_idx  = np.flatnonzero(is_nan)[0]
second_nan_idx = np.flatnonzero(is_nan)[1]
non_nan_idx = list(np.flatnonzero(~is_nan))
df = df.iloc[:, [first_nan_idx, second_nan_idx] + non_nan_idx]

# 4) Rename columns
df.columns = ["date", "hour"] + df.columns.tolist()[2:]

# 5) Create proper datetime with 15-minute intervals
# Get the date part
df["date_only"] = pd.to_datetime(df["date"]).dt.date

# Create 15-minute intervals: 00:15, 00:30, 00:45, 01:00, 01:15, etc.
# Each day has 96 intervals (24 hours * 4 intervals per hour)
intervals_per_day = 96
df["interval_in_day"] = df.index % intervals_per_day

# Convert interval number to time
# interval 0 = 00:15, interval 1 = 00:30, interval 2 = 00:45, interval 3 = 01:00, etc.
df["minutes_from_midnight"] = (df["interval_in_day"] + 1) * 15

# Create proper datetime
df["time"] = pd.to_datetime(df["date_only"].astype(str)) + pd.to_timedelta(df["minutes_from_midnight"], unit='minutes')

# 6) Remove temporary and old columns, put 'time' first
df = df.drop(columns=["date", "hour", "date_only", "interval_in_day", "minutes_from_midnight"])
df = df[["time"] + [c for c in df.columns if c != "time"]]

n_days = 10

df = df.iloc[:10]
prices_std = df['BTN A'].values
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from datetime import timedelta

def analyze_window(df,
                   col='BTN A',
                   start='10/01/2023',  # formato dd/mm/aaaa
                   days=2,
                   is_power_kw=False):
    """
    Filtra df pelo período [start, start+days), desenha gráfico e imprime consumo total.
    - df: DataFrame com colunas 'time' e a coluna de consumo/potência (ex: 'BTN A')
    - col: nome da coluna de interesse
    - start: string com data de início (dd/mm/aaaa)
    - days: número de dias a partir de start (inteiro)
    - is_power_kw: se True, interpreta a coluna como kW médios por 15 min e converte para kWh
    """

    # 1) Garantir datetime correto na coluna 'time' (formato do teu exemplo: "YYYY-MM-DD HH:MM:SS")
    ts = pd.to_datetime(df['time'], format='%Y-%m-%d %H:%M:%S', errors='coerce')

    # 2) Interpretar a data de início no formato PT (dd/mm/aaaa)
    start_dt = pd.to_datetime(start, dayfirst=True)
    end_dt = start_dt + timedelta(days=days)

    # 3) Filtrar janela [start, end)
    mask = (ts >= start_dt) & (ts < end_dt)
    if not mask.any():
        raise ValueError("Não há dados no intervalo pedido — confirma as datas.")

    # 4) Extrair série e garantir float
    s = pd.to_numeric(df.loc[mask, col], errors='coerce').astype(float)
    t = ts.loc[mask]

    # 5) Se a coluna estiver em kW médio por 15 min, converter para kWh do intervalo
    #    (energia = potência * 0,25 h)
    if is_power_kw:
        s_interval_kwh = s * 0.25
        y_label = 'kWh por intervalo (convertido de kW)'
    else:
        s_interval_kwh = s
        y_label = 'kWh por intervalo'

    # 6) Consumo total no período
    total_kwh = s_interval_kwh.sum()

    # 7) (Opcional) Consumo diário (útil para validação)
    serie = pd.Series(s_interval_kwh.values, index=t)
    consumo_diario = serie.resample('D').sum()

    # 8) Gráfico (15 min)
    plt.figure(figsize=(12, 4))
    plt.plot(t, s_interval_kwh.values)
    plt.title(f"{col} — {start_dt.strftime('%d/%m/%Y')} + {days} dia(s)\nTotal: {total_kwh:.3f} kWh")
    plt.xlabel('Tempo')
    plt.ylabel(y_label)
    plt.tight_layout()
    plt.show()

    # 9) Mostrar totais
    print(f"Período: [{start_dt} → {end_dt})")
    print(f"Consumo TOTAL no período: {total_kwh:.3f} kWh\n")
    print("Consumo diário (kWh):")
    print(consumo_diario.round(3))



#analyze_window(df, col='BTN A', start='1/03/2024', days=30, is_power_kw=False)

```

```python
import pandas as pd
from datetime import datetime


def _is_summer(date: datetime) -> bool:
    """Approximate check for Portuguese summer season (Apr-Oct)."""
    return 4 <= date.month <= 10


def _minutes(t) -> int:
    return t.hour * 60 + t.minute


def _period_daily(ts: datetime, option: str, season: str) -> str:
    m = _minutes(ts.time())
    if option == 'bi':
        if m >= 22 * 60 or m < 8 * 60:
            return 'vazio'
        return 'fora_vazio'
    if option == 'tri':
        if m >= 22 * 60 or m < 8 * 60:
            return 'vazio'
        if season == 'winter':
            if (8 * 60 + 30) <= m < (10 * 60 + 30) or (18 * 60) <= m < (20 * 60 + 30):
                return 'ponta'
            return 'cheias'
        else:  # summer
            if (10 * 60 + 30) <= m < (12 * 60) or (19 * 60 + 30) <= m < (21 * 60):
                return 'ponta'
            return 'cheias'
    return 'simples'


def _period_weekly(ts: datetime, option: str, season: str) -> str:
    m = _minutes(ts.time())
    wd = ts.weekday()  # Monday=0
    if option == 'bi':
        if wd <= 4:  # weekdays
            return 'vazio' if m < 7 * 60 else 'fora_vazio'
        if wd == 5:  # Saturday
            if season == 'winter':
                if m < 9 * 60 + 30 or 13 * 60 <= m < 18 * 60 + 30 or m >= 22 * 60:
                    return 'vazio'
                return 'fora_vazio'
            else:
                if m < 9 * 60 or 14 * 60 <= m < 20 * 60 or m >= 22 * 60:
                    return 'vazio'
                return 'fora_vazio'
        return 'vazio'  # Sunday
    if option == 'tri':
        if wd <= 4:  # weekdays
            if m < 7 * 60:
                return 'vazio'
            if season == 'winter':
                if 9 * 60 + 30 <= m < 12 * 60 or 18 * 60 + 30 <= m < 21 * 60:
                    return 'ponta'
                return 'cheias'
            else:
                if 9 * 60 + 15 <= m < 12 * 60 + 15:
                    return 'ponta'
                return 'cheias'
        if wd == 5:  # Saturday
            if season == 'winter':
                if 9 * 60 + 30 <= m < 13 * 60 or 18 * 60 + 30 <= m < 22 * 60:
                    return 'cheias'
                return 'vazio'
            else:
                if 9 * 60 <= m < 14 * 60 or 20 * 60 <= m < 22 * 60:
                    return 'cheias'
                return 'vazio'
        return 'vazio'  # Sunday
    return 'simples'


def get_tariff_period(ts: datetime, option: str, cycle: str) -> str:
    season = 'summer' if _is_summer(ts) else 'winter'
    if cycle == 'daily':
        return _period_daily(ts, option, season)
    return _period_weekly(ts, option, season)


def apply_indexed_tariff(prices_df: pd.DataFrame, vat_rate: float) -> pd.DataFrame:
    
    option = "bi"                         # simples, bi ou tri
    cycle = "daily"                       # daily ou weekly
    k1 = 1.28                             # multiplicador
    k2 = 0.0172                           # €/kWh
    losses = 0.085                        # 8,5 %
    rates = {
        "simples": 0.0600,
        "bi": {
            "fora_vazio": 0.0830,
            "vazio": 0.0149
        },
        "tri": {
            "ponta": 0.2469,
            "cheias": 0.0388,
            "vazio": 0.0149
        }
    }

    periods = []
    tariffs = []
    final_prices = []
    for ts, row in prices_df.iterrows():
        period = get_tariff_period(ts, option, cycle)
        periods.append(period)
        if option == 'simples':
            tar = rates['simples']
        elif option == 'bi':
            tar = rates['bi'][period]
        elif option == 'tri':
            tar = rates['tri'][period]
        else:
            tar = 0.0
            
        tariffs.append(tar)
        final_price = (row['price_omie_eur_kwh'] * (1 + losses) * k1 + k2 + tar) * (1 + vat_rate)
        print(final_price, row['price_omie_eur_kwh'])
        final_prices.append(final_price)

    prices_df['tariff_period'] = periods
    prices_df['tariff_energy_eur_kwh'] = tariffs
    prices_df['price_final_eur_kwh'] = final_prices
    return prices_df
```

```python
# %% [markdown]
# ### Preços REN Day-Ahead → €/kWh + Tarifas + Gráficos
# - Busca dados reais à API REN (DataHub) por dia
# - Converte para €/kWh com índice datetime (Hour 1..24 → 00..23)
# - Aplica a tua apply_indexed_tariff (k1, k2, perdas, períodos, IVA)
# - Plota Spot vs Final e a Diferença (Final − Spot)

# %%
import time
import requests
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime, timedelta

# ==========================
# 1) Funções de períodos/tarifa (o teu código)
# ==========================

def _is_summer(date: datetime) -> bool:
    """Approximate check for Portuguese summer season (Apr-Oct)."""
    return 4 <= date.month <= 10

def _minutes(t) -> int:
    return t.hour * 60 + t.minute

def _period_daily(ts: datetime, option: str, season: str) -> str:
    m = _minutes(ts.time())
    if option == 'bi':
        if m >= 22 * 60 or m < 8 * 60:
            return 'vazio'
        return 'fora_vazio'
    if option == 'tri':
        if m >= 22 * 60 or m < 8 * 60:
            return 'vazio'
        if season == 'winter':
            if (8 * 60 + 30) <= m < (10 * 60 + 30) or (18 * 60) <= m < (20 * 60 + 30):
                return 'ponta'
            return 'cheias'
        else:  # summer
            if (10 * 60 + 30) <= m < (12 * 60) or (19 * 60 + 30) <= m < (21 * 60):
                return 'ponta'
            return 'cheias'
    return 'simples'

def _period_weekly(ts: datetime, option: str, season: str) -> str:
    m = _minutes(ts.time())
    wd = ts.weekday()  # Monday=0
    if option == 'bi':
        if wd <= 4:  # weekdays
            return 'vazio' if m < 7 * 60 else 'fora_vazio'
        if wd == 5:  # Saturday
            if season == 'winter':
                if m < 9 * 60 + 30 or 13 * 60 <= m < 18 * 60 + 30 or m >= 22 * 60:
                    return 'vazio'
                return 'fora_vazio'
            else:
                if m < 9 * 60 or 14 * 60 <= m < 20 * 60 or m >= 22 * 60:
                    return 'vazio'
                return 'fora_vazio'
        return 'vazio'  # Sunday
    if option == 'tri':
        if wd <= 4:  # weekdays
            if m < 7 * 60:
                return 'vazio'
            if season == 'winter':
                if 9 * 60 + 30 <= m < 12 * 60 or 18 * 60 + 30 <= m < 21 * 60:
                    return 'ponta'
                return 'cheias'
            else:
                if 9 * 60 + 15 <= m < 12 * 60 + 15:
                    return 'ponta'
                return 'cheias'
        if wd == 5:  # Saturday
            if season == 'winter':
                if 9 * 60 + 30 <= m < 13 * 60 or 18 * 60 + 30 <= m < 22 * 60:
                    return 'cheias'
                return 'vazio'
            else:
                if 9 * 60 <= m < 14 * 60 or 20 * 60 <= m < 22 * 60:
                    return 'cheias'
                return 'vazio'
        return 'vazio'  # Sunday
    return 'simples'

def get_tariff_period(ts: datetime, option: str, cycle: str) -> str:
    season = 'summer' if _is_summer(ts) else 'winter'
    if cycle == 'daily':
        return _period_daily(ts, option, season)
    return _period_weekly(ts, option, season)

def apply_indexed_tariff(prices_df: pd.DataFrame, vat_rate: float) -> pd.DataFrame:
    """
    Espera 'price_omie_eur_kwh' e índice datetime.
    Produz:
      - tariff_period
      - tariff_energy_eur_kwh
      - price_final_eur_kwh
    """
    option = "bi"       # 'simples' | 'bi' | 'tri'
    cycle = "daily"     # 'daily' | 'weekly'
    k1 = 1.28           # multiplicador
    k2 = 0.0172         # €/kWh
    losses = 0.085      # 8,5 %

    rates = {
        "simples": 0.0600,
        "bi": {"fora_vazio": 0.0830, "vazio": 0.0149},
        "tri": {"ponta": 0.2469, "cheias": 0.0388, "vazio": 0.0149}
    }

    periods, tariffs, finals = [], [], []
    for ts, row in prices_df.iterrows():
        period = get_tariff_period(ts, option, cycle)
        periods.append(period)
        if option == 'simples':
            tar = rates['simples']
        elif option == 'bi':
            tar = rates['bi'][period]
        elif option == 'tri':
            tar = rates['tri'][period]
        else:
            tar = 0.0
        tariffs.append(tar)
        final_price = (row['price_omie_eur_kwh'] * (1 + losses) * k1 + k2 + tar) * (1 + vat_rate)
        finals.append(final_price)

    out = prices_df.copy()
    out['tariff_period'] = periods
    out['tariff_energy_eur_kwh'] = tariffs
    out['price_final_eur_kwh'] = finals
    return out

# ==========================
# 2) Fetch REN DataHub (dados reais)
# ==========================

REN_URL = "https://servicebus.ren.pt/datahubapi/electricity/ElectricityMarketPricesDaily"

def _extract_rows(json_obj):
    """A API pode devolver lista ou dict com 'Data'."""
    if isinstance(json_obj, list):
        return json_obj
    if isinstance(json_obj, dict):
        if 'Data' in json_obj and isinstance(json_obj['Data'], list):
            return json_obj['Data']
        # noutros casos pode já ser uma lista em alguma chave
        for v in json_obj.values():
            if isinstance(v, list) and v and isinstance(v[0], dict):
                return v
    return []

def fetch_ren_prices_range(start_date: str, end_date: str, culture: str = "pt-PT", pause_s: float = 0.15) -> pd.DataFrame:
    """
    Vai buscar preços Day-Ahead à REN para [start_date, end_date] (YYYY-MM-DD).
    Devolve DataFrame com colunas: Date, Hour, Price_PT_€/MWh (ou equivalente),
    mais 'datetime' e 'price_omie_eur_kwh'.
    """
    start = datetime.fromisoformat(start_date).date()
    end = datetime.fromisoformat(end_date).date()
    if end < start:
        raise ValueError("end_date < start_date")

    frames = []
    sess = requests.Session()
    d = start
    while d <= end:
        params = {"culture": culture, "date": d.isoformat()}
        r = sess.get(REN_URL, params=params, timeout=15)
        r.raise_for_status()
        rows = _extract_rows(r.json())
        if not rows:
            print(f"[WARN] Sem dados para {d}")
        else:
            df_day = pd.DataFrame(rows)

            # Detectar coluna de preço em €/MWh (tem 'Price' e 'MWh' no nome)
            price_col = None
            for c in df_day.columns:
                cl = c.lower()
                if "price" in cl and "mwh" in cl:
                    price_col = c
                    break
            if price_col is None:
                raise ValueError(f"Não encontrei coluna de preço €/MWh em {list(df_day.columns)}")

            # Normalizar colunas
            df_day = df_day.rename(columns={
                'Date': 'Date',
                'Hour': 'Hour',
                price_col: 'Price_PT_€/MWh'
            })

            # Construir datetime (Hour 1..24 -> 00..23)
            df_day["datetime"] = pd.to_datetime(df_day["Date"]) + pd.to_timedelta(df_day["Hour"] - 1, unit="h")

            # €/MWh → €/kWh
            df_day["price_omie_eur_kwh"] = df_day["Price_PT_€/MWh"] / 1000.0

            frames.append(df_day)

        time.sleep(pause_s)
        d += timedelta(days=1)

    if not frames:
        raise RuntimeError("Sem dados no intervalo.")
    df = pd.concat(frames, axis=0).sort_values("datetime").set_index("datetime")
    return df

# ==========================
# 3) Executar: buscar, tarifar, plotar
# ==========================

# Ajusta aqui o intervalo e o IVA:
START = "2025-07-25"
END   = "2025-07-26"
VAT   = 0.23

# Buscar dados reais (no formato que mostraste)
raw_df = fetch_ren_prices_range(START, END, culture="pt-PT")

# Aplicar a tua tarifa indexada
priced = apply_indexed_tariff(raw_df, vat_rate=VAT)

# Ver topo para confirmar
display_cols = ["Date", "Hour", "Price_PT_€/MWh", "price_omie_eur_kwh", "tariff_period", "tariff_energy_eur_kwh", "price_final_eur_kwh"]
display(priced[display_cols].head(10))

# Gráfico 1 — Spot vs Final
plt.figure(figsize=(12,6))
plt.plot(priced.index, priced["price_omie_eur_kwh"], label="Preço Spot (OMIE)")
plt.plot(priced.index, priced["price_final_eur_kwh"], label="Preço Final (com perdas+taxas+IVA)")
plt.title(f"Spot vs Final (€/kWh)  [{START} → {END}]")
plt.xlabel("Data/Hora")
plt.ylabel("€/kWh")
plt.legend()
plt.grid(True, linestyle="--", alpha=0.5)
plt.tight_layout()
plt.show()

# Gráfico 2 — Diferença (Final − Spot)
diff = priced["price_final_eur_kwh"] - priced["price_omie_eur_kwh"]
plt.figure(figsize=(12,4))
plt.plot(diff.index, diff.values, label="Δ (Final − Spot)")
plt.title("Diferença entre preço final e spot (€/kWh)")
plt.xlabel("Data/Hora")
plt.ylabel("€/kWh")
plt.grid(True, linestyle="--", alpha=0.5)
plt.tight_layout()
plt.show()
```
